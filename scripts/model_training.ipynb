```python
from sklearn.feature_extraction.text import TfidfVectorizer
import spacy
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
spacy_model = spacy.load('en_ner_bionlp13cg_md')
tv = TfidfVectorizer()
```

```python
def lemm(x, model):
    """ Lemmatize each word in a sentence"""

    clean = []
    tokens = model(x)
    for token in tokens:
        if not token.is_stop:
            clean.append(token.lemma_)
    return " ".join(clean)
```

```python
df['tokens'] = df['title'].apply(lemm, model=spacy_model)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['tokens'], df['cit_per_yr'])
```

```python
example = tv.fit_transform(X_train)
```

```python
df_tv = pd.DataFrame(example.todense(), columns = tv.get_feature_names()) # dense because we save zeros
```

```python
df_tv
```

```python
rf = RandomForestRegressor()
rf_param = { 'max_depth': [1, 5, 8, 32],#, 8, 32, 128, None],
    'n_estimators': [10, 100]}#, 100, 1000, 1000]}

grid_rf= GridSearchCV(rf,rf_param,scoring = 'neg_mean_squared_error',cv=5)
grid_rf.fit(df_tv, y_train)
ypred = grid_rf.predict(df_tv)
print (grid_rf.best_params_)


best_score = (grid_rf.best_score_)*-1
print('Mean Square Error is : ',best_score)
```
