```python
df_1 = pd.read_csv('../biochemistry_meta_short.csv')
```

```python
df_1['journal'] = df_1['journal'].str.strip('[]')
df_1['journal'] = df_1['journal'].str.strip("'")
```

```python
df
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['year'] = df.date.dt.year
```

```python
df['yrs_passed'] = 2020 - df.year + 1
```

```python
df['cit_per_yr'] = df['citation_count'] / df['yrs_passed']
```

```python
df
```

## Try with journal name and title TFIDF 

```python
df_train_pre = df[['title', 'journal', 'cit_per_yr']]
```

```python
df_train_pre
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tv = TfidfVectorizer()
```

```python
example = tv.fit_transform(df_train_pre['title'])
```

```python
example
```

```python
df_tv = pd.DataFrame(example.todense(), columns = tv.get_feature_names()) # dense because we save zeros
```

```python
df_tv
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_tv, df_train_pre['cit_per_yr'], random_state=42)
```

### No need to scale, they're all in between 0-1

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(random_state=42)
```

```python
pca.fit(X_train, y_train)
```

```python
T = pd.DataFrame(pca.transform(X_train))
```

```python
T.head()
```

```python
# Plot the explained variance by all the features
pd.DataFrame(pca.explained_variance_ratio_).plot.bar() #!!!!
```

```python
import numpy as np
pd.Series(np.cumsum(pca.explained_variance_ratio_)).plot()
```

```python
pca90 = PCA(n_components=.9)
```

```python
pca90.fit(X_train)
```

```python
pca90.explained_variance_ratio_
```

```python
W90 = pd.DataFrame(pca90.components_, columns=df_tv.columns)
```

```python
W90
```

### Scale y data

```python
ytrain_sc = np.log1p(y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
rf = RandomForestRegressor()
rf_param = { 'max_depth': [1, 5, 8, 32],#, 8, 32, 128, None],
    'n_estimators': [10, 100]}#, 100, 1000, 1000]}

grid_rf= GridSearchCV(rf,rf_param,scoring = 'neg_mean_squared_error',cv=5)
T_train = pd.DataFrame(pca90.transform(X_train))
grid_rf.fit( T_train, ytrain_sc)
ypred = grid_rf.predict(T_train)
print (grid_rf.best_params_)


best_score = (grid_rf.best_score_)*-1
print('Mean Square Error is : ',best_score)
```
