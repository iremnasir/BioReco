```python
def run_models(abstract):
    """
    Runs two Spacy models to recognize NER
    Returns bionlp and bc5cdr output
    """
    bio_abst = ner_bio(abstract)
    bc_abst = ner_bc(abstract)
    
    return bio_abst, bc_abst
```
